```python
#Importing the required modules
import numpy as np 
import pandas as pd 
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('World_Bank_Data_India.csv')
data
```

```python
fig, axes1 = plt.subplots(figsize=(20,8))
plt.plot(data['Years'],data['POP_TOTL'], '*-')
plt.plot(data['Years'],data['POP_014'],'*-' )
plt.plot(data['Years'],data['POP_1564'], '*-')
plt.plot(data['Years'],data['POP_65'], '*-')
plt.legend(prop={'size': 16})
```

There has been continous increase in the population and the maximum increase we see is in the age group 15-64.

```python
plt.subplots(figsize=(20,3))
plt.plot(data['Years'],data['EMP_TOTL'], '*-')
plt.title("Total Employment growth")

fig, axes1 = plt.subplots(figsize=(20,6))
#plt.plot(data['Years'],data['EMP_TOTL'], '*-')
plt.plot(data['Years'],data['EMP_SELF'],'*-' ,label="Self Employed")
plt.plot(data['Years'],data['EMP_SRV'], '-',label="Service Sector")
plt.plot(data['Years'],data['EMP_IND'], '*-',label="Industrial Sector")
plt.plot(data['Years'],data['EMP_AGR'], '*-',label="Agriculture Sector")
plt.title("Employment Growth Sector wise")
plt.legend(prop={'size': 16})
```

We can see that overall employment of the country has increased, and the max increase we see is in the Service and Industrial sector which hold true as there has been tremendous development in technology  in last 2 decades.

```python
fig, axes1 = plt.subplots(figsize=(20,4))
plt.plot(data['Years'],data['GDP_IND'], '*-')

plt.plot(data['Years'],data['GDP_AGR'],'*-' )
plt.legend(prop={'size': 16})
```

The contribution of Agriculture sector in Total GDP of country is decreasing while that of Industrial Sector remain same.

Thus it can be inferred that contribution of Service sectoris only growing

```python
fig, axes1 = plt.subplots(figsize=(20,4))
plt.plot(data['Years'],data['FRTL'], '*-')
plt.plot(data['Years'],data['BR'],'*-' )
plt.plot(data['Years'],data['DR'],'*-' )
plt.legend(prop={'size': 16})
```

Fertility rate of the country has remained the same in last 3 decades but birth rate has decreased in this time frame thus a conclusion cant be argued that women during her labor days were provided with good medical facilities and she lived until the delivery of the child.

```python
# Convert fractional indicators to actual numbers.
data['UEM_TOTL']= data['UEM_TOTL']*data['LF_TOTL']/100
data['LF_BASC_EDU']= data['LF_BASC_EDU']*data['LF_TOTL']/100
data['LF_INTM_EDU']= data['LF_INTM_EDU']*data['LF_TOTL']/100
data['LF_ADVN_EDU']= data['LF_ADVN_EDU']*data['LF_TOTL']/100
data['EMP_SRV']= data['EMP_SRV']*data['EMP_TOTL']/100
data['EMP_IND']= data['EMP_IND']*data['EMP_TOTL']/100
data['EMP_AGR']= data['EMP_AGR']*data['EMP_TOTL']/100
data['GDP_SRV']=100-data['GDP_AGR']-data['GDP_IND']
data.head()
```

```python
fig, axes2 = plt.subplots(figsize=(10,8))

axes2.plot(data['Years'],data['POP_1564'],color='red',lw=4)
axes2.plot(data['Years'],data['LF_TOTL'],color='green',lw=3)
axes2.plot(data['Years'],data['EMP_TOTL'],color='blue',lw=2)
axes2.plot(data['Years'],data['UEM_TOTL'],color='black',lw=2)
axes2.fill_between(data['Years'], data['POP_1564'], data['LF_TOTL'], color="yellow", alpha=0.5)
axes2.set_title("Employment rate for Last 26 years vs Population rise")
axes2.set_xlabel("Years")
axes2.set_ylabel("Population in Billions")
axes2.legend(loc=0)
```

According to World Bank "Labor force comprises people ages 15 and older who supply labour for the production of goods and services during a specified period. It includes people who are currently employed and people who are unemployed but seeking work as well as first-time job-seekers."

So by that definition, we can summarise about the increasing gap between Population of that age and Labour force. The reason might be due to increasing number of students of 16+ before seeking employment as they are not being considered within labour force or it might also be that educated family workers are growing. Whatever the reason might be the summary we get here is that we are failing to utilise our full labour force capability for our economy.

It also seems that Unemployment has not increased since Globalisation and has stayed the same.
  

```python
fig, axes3 = plt.subplots(figsize=(15,8))

axes3.plot(data['Years'],data['LF_BASC_EDU'],color='#008B8B',lw=3, marker='o', markersize=10)
axes3.plot(data['Years'],data['LF_INTM_EDU'],color='#0000ff',lw=3, marker='p', markersize=10)
axes3.plot(data['Years'],data['LF_ADVN_EDU'],color='#00EEEE',lw=3, marker='s', markersize=10)
axes3.plot(data['Years'],data['EMP_SRV'],color='#ff0000',lw=4)
axes3.plot(data['Years'],data['EMP_IND'],color='#660000',lw=4)
axes3.plot(data['Years'],data['EMP_AGR'],color='#cc0000',lw=4)
axes3.set_title("Rate of Education level of our Labur force and their sector wise Employment")
axes3.set_xlabel("Years")
axes3.set_ylabel("Population in 100 millions")
axes3.legend(loc=0)
```

According to the International Standard Classification of Education 2011, Basic education comprises primary education or lower secondary education, Intermediate education includes upper secondary or post-secondary non tertiary education, Advanced education includes short-cycle tertiary education, a bachelors degree or equivalent education level, a masters degree or equivalent education level, or doctoral degree or equivalent education.

We conclude that more people are acquiring minimum advanced educational qualification and a population of basic and intermediate education qualification labours are gradually decreasing. This shows that our labour force is getting much more advanced educated. But does our economy have that space to contain this massive army of educated labour? We can clearly see that only service sector is filling that gap, but this will create a vast educated unemployment in future (which doesn't show up in general unemployment data) as most of the service employment does not require highly advanced educational qualification.

We can also observe that since 2010, the population in industrial sector is decreasing which is due to high educated but low skilled labour population, not suitable for Industry.

We also see the sharp decrease in agricultural sector since 2005. Is it due to bad governance of our agricultural sector or agricultural sectors are getting more automated and less manual?

We also see a sudden increase in service sector since 2010. Is it due to the booming effect of the internet and cellular phones? So large number of the seeking labour force are entering this sector due to sudden vast varieties of service demand everywhere

```python
fig, axes4 = plt.subplots(figsize=(15,8))

axes4.plot(data['Years'],data['GDP_SRV'],color='red',lw=4)
axes4.plot(data['Years'],data['GDP_IND'],color='blue',lw=4)
axes4.plot(data['Years'],data['GDP_AGR'],color='brown',lw=4)
axes4.set_title("Rate of GDP sector wise")
axes4.set_xlabel("Years")
axes4.set_ylabel("% of total GDP")
axes4.legend(loc=0)
```

The above conclusion can be backed with Rate of GDP for each sector wise. We see the slow GDP decrease in the industrial sector and a very significant decrease in the agriculture sector and very substantial increase in the service sector which is currently accounting for more than 50% of GDP.

```python
fig, axes5 = plt.subplots(1,4,figsize=(20,8))

axes5[0].plot(data['Years'],data['GDP_PP_EMP'],color='green',lw=3)
axes5[0].set_title("GDP per person employed")
axes5[1].plot(data['Years'],data['INFL'],color='blue',lw=3)
axes5[1].set_title("Inflation, consumer prices")
axes5[2].plot(data['Years'],data['XPD_TOTL_EDU'],color='black',lw=3)
axes5[2].set_title("Government expenditure on education")
axes5[3].plot(data['Years'],data['FRTL'],color='Orange',lw=3)
axes5[3].set_title("Fertility rate, total (births per woman)")
```

We can summarise that even though our rate of employment is not increasing as before 2010, the rise of GDP per employed person has been excellent since 2010 of which service sector contributes the most. That's a good sign for our economy. But what about society? Is this sustainable?

From Inflation, we can not conclude anything accurately, but its sure that our government has failed to control inflation as there are steep highs and lows.

The sad news is, even after such a reasonable increase in GDP our government is not expending that much on education. Also though there was a good sign after 2010, that too has started declining. That's a bad sign for us as a society. It creates Not only low skilled labour but also a significant division within employers concerning self-sustainability. This will generate a lot of panic situation within our community.

The decrease in fertility rate confirms two reasons behind this, first the increase in the number of women giving fewer births and second infertility.

```python
fig, axes6 = plt.subplots(1,2,figsize=(20,10))

axes6[0].plot(data['Years'],data['BR'],color='green',lw=3)
axes6[0].set_title("Birth Rate per 1000 people")
axes6[0].plot(data['Years'],data['DR'],color='red',lw=3)
axes6[0].set_title("Death Rate")
axes6[0].set_xlabel("Years")
axes6[0].set_ylabel("Person per 1000")
axes6[0].legend(loc=0)
axes6[1].plot(data['Years'],data['D_COM'],color='#ff0000',lw=3, marker='o', markersize=10)
axes6[1].plot(data['Years'],data['D_INJ'],color='#660000',lw=3, marker='p', markersize=10)
axes6[1].plot(data['Years'],data['D_NC'],color='#cc0000',lw=3, marker='s', markersize=10)
axes6[1].set_title("Death Rate due to different Reasons")
axes6[1].set_xlabel("Years")
axes6[1].set_ylabel("% of Death Rate")
axes6[1].legend(loc=0)
```

We can clearly see that Birth rate is decreasing at a much faster rate than Death Rate. That's ALARMING.

Also number of Deaths due to Non-Communicable (cancer, diabetes mellitus, cardiovascular diseases, digestive diseases, skin diseases, musculoskeletal diseases, and congenital anomalies) and Communicable diseases (parasitic diseases, respiratory infections, and nutritional deficiencies) were same during beginning of millennium but within last 15 years Non communicable diseases have taken a significant role. That pretty much clears out that our medical facilities have developed a lot minimising communicable diseases, but our present lifestyle has given room for non-communicable diseases to grow.
